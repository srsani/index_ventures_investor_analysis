# Downloading datasets:

The datasets are stored in PostgreSQL database in amazon RDS. So I used psycopg2 to connect to the server and pandas to download the databases in to a datafram and then save them to  local drive for future use.

```python
import pandas as pd
import psycopg2
from IPython.display import display, Audio
from IPython.display import HTML

from pandas_summary import DataFrameSummary
```

```python
host= open('host.txt', 'r').read()
password= open('password.txt', 'r').read()
user= open('user.txt', 'r').read()
dbname= open('dbname.txt', 'r').read()
port= open('port.txt', 'r').read()
```

```python
try:
    conn = psycopg2.connect(dbname=dbname , user=user, host=host, password=password, port = '5432')
    cur = conn.cursor()
except:
    print ("I am unable to connect to the database")
    
```

```python
cur.execute("select relname from pg_class where relkind='r' and relname !~ '^(pg_|sql_)';")
table_list = cur.fetchall()
table_list
```

```python
tables = [pd.read_sql('select * from '+table_list[i][0]+';', conn) for i in range(len(table_list))]
```

# Looking at the Summary of each datafram:

```python
for t in tables: 
    try: display(DataFrameSummary(t).summary())
    except : print("Dataset has non hashable objects")
```

# Saveing each datafram to a .csv:

```python
tb_name = []
dflist = []
for i in range(len(table_list)):
    tb_name.append(table_list[i][0])
    
for i,val in enumerate(tb_name):
    globals()[val] = tables[i]
    dflist.append(globals()[val])

for i in range(len(table_list)): 
    df = dflist[i]
    df.to_csv('./data2/'+tb_name[i]+'.csv')
```
